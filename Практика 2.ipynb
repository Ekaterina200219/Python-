# Задание 3 Напишите функцию которая получает на вход целое число и вычисляет сумму цифр из которых это число состоит (1872 = 1+8+7+2) 

```python
def integer(number):
    summ = 0
    while number > 0:
        digit = number % 10 #нахождение остатка от деления на 10
        summ = summ + digit
        number = number // 10 #деление нацело на 10
    return summ
```

```python
print(integer(25894))
```

# Задание 4 Нам необходимо понять нет ли у нас неповторяющихся в производстве на разных заводах товаров. Напишите программу которая создает 2 множества, заполняет их случайным количеством “товаров” (рыба,мясо,сталь,игрушки и т.д из заранее сформированного вами списка (не меньше 15 “товаров”) а после сравнивает их и на выход поступают те что не повторяются на обоих. . 

```python
import random 

products = ["сливы", "абрикосы", "яблоки", "ананас", "мандарины", "персики", "бананы", "картошка", "манго", "помидоры", "огурцы",
            "укроп", "мясо", "рыба", "орехи"]
empty_1 = set() #пустое множество
empty_2 = set()
while True:
    for i in range(random.randint(1, 15)):
        empty_1.add(products[random.randint(0, len(products) - 1)])
    for i in range(random.randint(1, 15)):
        empty_2.add(products[random.randint(0, len(products) - 1)])
    if empty_1.isdisjoint(empty_2): #Производит проверку, не пересекается ли рассматриваемое множество с заданным.
        print(empty_1, empty_2)
        break
    else:
        empty_1.clear()
        empty_2.clear()
```
